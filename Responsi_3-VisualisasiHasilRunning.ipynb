```python
# Import Library yang diperlukan
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.model_selection import cross_val_predict, train_test_split
import scikitplot as skplt
```

```python
# Melakukan Pre-Processing
Dataset = pd.read_csv("C:\\Users\\Muhammad Raihan\\Downloads\\Obfuscated-MalMem2022.csv")
X = Dataset.drop(['Category','Class'],axis=1).values    #Di drop karena algoritma klasifikasi tidak menerima variabel tipe object
y = Dataset['Class'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
# Melakukan Klasifikasi binary atau 2 class untuk dataset CIC-MalMem-2020 menggunakan Gradient Boosting dan Random Forest
gbc = GradientBoostingClassifier()
gbc_model = gbc.fit(X_train, y_train)
rfc = RandomForestClassifier()
rfc_model = rfc.fit(X_train, y_train)
```

```python
# melakukan prediksi kelas dan probabilitas kelas dari data uji (X_test)
y_gbc_pred = gbc_model.predict(X_test)
y_gbc_proba = gbc_model.predict_proba(X_test)
y_rfc_pred = rfc_model.predict(X_test)
y_rfc_proba = rfc_model.predict_proba(X_test)
```

```python
# Melakukan visualisasi Confusion Matrix Gradient Boosting
skplt.metrics.plot_confusion_matrix(y_test, y_gbc_pred, normalize=True, title = 'Confusion Matrix untuk GBC')
```

```python
# Melakukan visualisasi Confusion Matrix Random Forest
skplt.metrics.plot_confusion_matrix(y_test, y_rfc_pred, normalize=True, title = 'Confusion Matrix untuk RFC')
```

```python
# Melakukan Visualisasi Receiver Operating Characteristic untuk Gradient Boosting
skplt.metrics.plot_roc(y_test, y_gbc_proba, title = 'ROC Plot untuk GBC')
```

```python
# Melakukan Visualisasi Receiver Operating Characteristic untuk Random Forest
skplt.metrics.plot_roc(y_test, y_rfc_proba, title = 'ROC Plot untuk RFC')
```

```python
# Melakukan Visualisasi Precision Recall Curve untuk Gradient Boosting
skplt.metrics.plot_precision_recall(y_test, y_gbc_proba, title = 'PR Curve for GBC')
```

```python
# Melakukan Visualisasi Precision Recall Curve untuk Random Forest
skplt.metrics.plot_precision_recall(y_test, y_rfc_proba, title = 'PR Curve for RFC')
```

```python
# Ubah target (y_test) menjadi numerik 0 dan 1 karena hanya bisa dilakukan pada binary classification
le = LabelEncoder()
y_test_numeric = le.fit_transform(y_test)

# Melakukan Visualisasi Calibration Plot
probas_list = [y_gbc_proba, y_rfc_proba]
clf_names = ['GBC', 'RFC']
skplt.metrics.plot_calibration_curve(y_test_numeric, probas_list=probas_list, clf_names=clf_names)
```

```python
# Melakukan Visualisasi Cumulative Gains Curve untuk Gradient Boosting
skplt.metrics.plot_cumulative_gain(y_test, y_gbc_proba, title = 'Cumulative Gains Chart for GBC')
```

```python
# Melakukan Visualisasi Cumulative Gains Curve untuk Random Forest
skplt.metrics.plot_cumulative_gain(y_test, y_rfc_proba, title = 'Cumulative Gains Chart for RFC')
```

```python
# Melakukan Visualisasi Lift Curve untuk Gradient Boosting
skplt.metrics.plot_lift_curve(y_test, y_gbc_proba, title = 'Lift Curve for GBC')
```

```python
# Melakukan Visualisasi Lift Curve untuk Random Forest
skplt.metrics.plot_lift_curve(y_test, y_rfc_proba, title = 'Lift Curve for RFC')
```
